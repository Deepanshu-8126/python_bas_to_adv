Even and Odd Checker

```python
# check a number is odd or even
n=int(input("Enter an number"))
if n%2==0:
  print("Even number")
else:
  print("Odd number")
```

```python
# factorial number
n=int(input("Enter number "))
fact=1
for i in range(1,n+1):
   fact=fact*i
print(fact)
```
